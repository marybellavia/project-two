```python
# Dependencies and Setup
import os
import pandas as pd
import requests
import json
```

```python
#read the data sets from Resources file
csv_path = "static/data/rent_cleaned.csv"
rent_df = pd.read_csv(csv_path)

rent_df
```

```python
# filtering down to the month and year of interest
df = rent_df.loc[(rent_df["Year"] == 2020) & (rent_df["Month"] == 1)]
```

```python
df
```

```python
# getting the percentile
percentile = round(len(df["Price"].index) * .8)
```

```python
# calculation to get the numberal value for to determine the top 20%
sorted(list(df["Price"].values))[percentile]
```
